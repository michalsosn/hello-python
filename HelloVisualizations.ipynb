# An image

```python
def show_image(image):
    fig = plt.figure(figsize=(12, 12))
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
image = misc.face(gray = True)
show_image(image)
```

```python
afilter = np.ones((20, 20)) / 400
filtered = signal.convolve2d(image, afilter)
show_image(filtered)
```

```python
bfilter = np.array([[0, 1, 2], [-1, 0, 1], [-2, -1, 0]])
filtered = signal.convolve2d(image, bfilter)
show_image(filtered)
```
